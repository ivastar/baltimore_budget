```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
budget = pandas.read_csv('baltimore_budget.csv')
budget_sorted = budget[:-1].sort_values(by='2020',ascending=False)
budget_sorted['2020'] = budget_sorted['2020'] /1e6
budget_sorted['2019'] = budget_sorted['2019'] /1e6
budget_sorted['Difference'] = (budget_sorted['2020'] - budget_sorted['2019']).astype('int')
budget_sorted['Percent_diff'] = 100*((budget_sorted['2020'] - budget_sorted['2019'])/budget_sorted['2019'])
budget_sorted['Fraction_of_total'] = 100*budget_sorted['2020']/3071
```

```python
f, [ax1, ax2] = plt.subplots(ncols=2, nrows=1,figsize=(12, 6))
f.suptitle('Baltimore City 2020 Budget')
sns.set_color_codes("pastel")
sns.set(style="whitegrid")
sns.barplot(x="2020", y="Agency", data=budget_sorted.sort_values(by='2020', ascending=False)[:10],
            label="Million US Dollars", palette=["gray",'b',"gray","green","gray","gray","gray","gray","gray","gray"], orient='h', alpha=0.7, ax=ax1)
ax1.set_xlabel('Agency Budget [Million US Dollars]')
ax1.set_ylabel('')

sns.barplot(x="Fraction_of_total", y="Agency", data=budget_sorted.sort_values(by='2020', ascending=False)[:10],
            label="Million US Dollars", palette=["gray",'b',"gray","green","gray","gray","gray","gray","gray","gray"], orient='h', alpha=0.7, ax=ax2)
ax2.set_xlabel('Percent of Total Budget')
ax2.set_yticklabels('')
ax2.set_ylabel('')
```

```python
f, [ax1,ax2] = plt.subplots(ncols=2, nrows=1,figsize=(12, 6))
f.suptitle('Baltimore City 2019 to 2020 Budget Changes')
sns.set_color_codes("pastel")
sns.set(style="whitegrid")

sns.barplot(x="Difference", y="Agency", data=budget_sorted.sort_values(by=['Difference','2020'], ascending=False)[:21],
            label="Million US Dollars", palette=['b']+["gray"]*20, orient='h', alpha=0.7, ax=ax1)
ax1.set_xlabel('Increace 2019 to 2020 [Million US Dollars]')
ax1.set_ylabel('')

sns.barplot(x="Percent_diff", y="Agency", data=budget_sorted.sort_values(by=['Difference','2020'], ascending=False)[:21],
            label="Million US Dollars", palette=['b']+["gray"]*20, orient='h', alpha=0.7, ax=ax2)
ax2.set_xlabel('Percent Increase 2019 to 2020')
ax2.set_yticklabels('')
ax2.set_ylabel('')
```

```python
police_budget = pandas.read_csv('police_budget.csv')
police_budget_sorted = police_budget[:-1].sort_values(by='2020',ascending=False)
police_budget_sorted['2020'] = police_budget_sorted['2020'] /1e6
police_budget_sorted['2019'] = police_budget_sorted['2019'] /1e6
police_budget_sorted['Difference'] = police_budget_sorted['2020']-police_budget_sorted['2019']
police_budget_sorted['Percent_Difference'] = 100*(police_budget_sorted['2020']-police_budget_sorted['2019'])/police_budget_sorted['2019']
police_budget_sorted['Percent_of_total'] = 100*police_budget_sorted['2020']/536
```

```python
f, [ax1,ax2] = plt.subplots(ncols=2, nrows=1,figsize=(12, 6))
f.suptitle('Baltimore City Police Department 2020 Budget')
sns.set_color_codes("pastel")
sns.set(style="whitegrid")

sns.barplot(x="2020", y="Police_item", data=police_budget_sorted.sort_values(by=['2020'], ascending=False),
            label="Million US Dollars", color='b', orient='h', alpha=0.7, ax=ax1)
ax1.set_xlabel('2020 Budget [Million US Dollars]')
ax1.set_ylabel('')

sns.barplot(x="Percent_of_total", y="Police_item", data=police_budget_sorted.sort_values(by=['2020'], ascending=False),
            label="Million US Dollars", color='b', orient='h', alpha=0.7, ax=ax2)
ax2.set_xlabel('Percent of Total Police Budget')
ax2.set_yticklabels('')
ax2.set_ylabel('')
```

```python
f, [ax1,ax2] = plt.subplots(ncols=2, nrows=1,figsize=(12, 6))
f.suptitle('2019 to 2020 Baltimore City Police Department Budget Changes')
sns.set_color_codes("pastel")
sns.set(style="whitegrid")

sns.barplot(x="Difference", y="Police_item", data=police_budget_sorted.sort_values(by=['2020'], ascending=False),
            label="Million US Dollars", color='b', orient='h', alpha=0.7, ax=ax1)
ax1.set_xlabel('2019 to 2020 Increase [Million US Dollars]')
ax1.set_ylabel('')

sns.barplot(x="Percent_Difference", y="Police_item", data=police_budget_sorted.sort_values(by=['2020'], ascending=False),
            label="Million US Dollars", color='b', orient='h', alpha=0.7, ax=ax2)
ax2.set_xlabel('Percent Increase 2019 to 2020')
ax2.set_yticklabels('')
ax2.set_ylabel('')
```
