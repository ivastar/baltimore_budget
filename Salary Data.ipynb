```python
import pandas
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
salaries = pandas.read_csv('Baltimore_City_Employee_Salaries_FY2019.csv')
```

```python
salaries.sort_values(by='Gross',ascending=False)
```

```python
salaries["DESCR"]= salaries["DESCR"].str.split("(", n = 1, expand = True)
salaries["DESCR"]= salaries["DESCR"].str.strip()
```

```python
salaries[salaries['JOBTITLE'] == 'Mayor']
```

```python
mayor_salary = salaries[salaries['JOBTITLE'] == 'Mayor']['Gross'].values[0]
```

```python
salaries.DESCR.unique()
```

```python
police = salaries[salaries['DESCR'] == 'Police Department']
all_but_police = salaries[salaries['DESCR'] != 'Police Department']
```

```python
police
```

```python
print('median non-police: ', np.nanmedian(all_but_police['ANNUAL_RT']))
print('median police: ', np.nanmedian(police['ANNUAL_RT']))
print('mayor: ', salaries[salaries['JOBTITLE'] == 'Mayor']['ANNUAL_RT'].values[0])
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.distplot(all_but_police['ANNUAL_RT'],kde = False, color='gray')
police['ANNUAL_RT'].hist(bins=30, alpha=0.6)
#ax.plot([mayor_salary,mayor_salary],[0,1500], color='red')
ax.plot([np.nanmedian(all_but_police['ANNUAL_RT']),np.nanmedian(all_but_police['ANNUAL_RT'])],[0,2000], color='gray', label='Non-Police')
ax.plot([np.nanmedian(police['ANNUAL_RT']),np.nanmedian(police['ANNUAL_RT'])],[0,2000], color='lightblue', label='Police')
ax.set_ylim(0,1700)
ax.set_xlabel('Annual Rate')
ax.set_ylabel('Number of Employees')
```

```python
print('median non-police: ', np.nanmedian(all_but_police['Gross']))
print('median police: ', np.nanmedian(police['Gross']))
print('mayor: ', mayor_salary)
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.distplot(all_but_police['Gross'],kde = False, color='gray')
police['Gross'].hist(bins=30, alpha=0.6)
ax.plot([mayor_salary,mayor_salary],[0,600], color='red')
ax.plot([np.nanmedian(all_but_police['Gross']),np.nanmedian(all_but_police['Gross'])],[0,800], 
        color='gray', label='Non-Police')
ax.plot([np.nanmedian(police['Gross']),np.nanmedian(police['Gross'])],[0,700], color='lightblue', label='Police')
ax.set_ylim(0,900)
ax.set_xlabel('Gross 2019 Salary')
ax.set_ylabel('Number of Employees')
ax.text(55000, 810, 'Non-police\nMedian:\n$51,258', color='gray', va='center')
ax.text(95000, 710, 'Police\nMedian:\n$93,278', color='blue', va='center', alpha=0.6)
ax.text(131000, 610, 'Mayor:\n$127,985', color='red', va='center')
```

```python
len(police[police['Gross']>mayor_salary])
```

```python
print('Fraction of police officers paid more than the mayor: ', len(police[police['Gross']>mayor_salary])/len(police))
```

```python
print('Fraction of other city employees paid more than the mayor: ', len(all_but_police[all_but_police['Gross']>mayor_salary])/len(all_but_police))
```

```python
police['Overpay'] = 100*(police['Gross'] - police['ANNUAL_RT'])/police['ANNUAL_RT']
police.sort_values(by='Gross',ascending=False)[:10]
```

```python
f, ax = plt.subplots(figsize=(10, 7))

sns.barplot(x="Gross", y="NAME", data=police.sort_values(by='Gross',ascending=False)[:20],
            label="Overtime", color="lightblue")

sns.barplot(x="ANNUAL_RT", y="NAME", data=police.sort_values(by='Gross',ascending=False)[:20],
            label="Base", color="blue", alpha=0.5)

# Add a legend and informative axis label
ax.legend(ncol=1, loc="lower right", frameon=True)
#ax.set(xlim=(0, 24), ylabel="",
#       xlabel="Automobile collisions per billion miles")
sns.despine(left=True, bottom=True)
ax.set_ylabel('')
ax.set_xlabel('Gross 2019 Salary')
```
